```python
import numpy as np
def plot_images(images):

    rows = 2
    cols = 4

    fig = plt.figure(figsize = (10, 6))

    for i in range(rows*cols):

        ax = fig.add_subplot(rows, cols, i+1)
        
        image = images[i]

        ax.imshow(image,cmap='gray')
        ax.set_title(i)
        ax.axis('off')
```
